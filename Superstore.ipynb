```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%m/%d/%Y')
df.info()
```

```python
# Find the single biggest loss-making order
worst_order = df[df['Profit'] == df['Profit'].min()]
print(worst_order[['Order ID', 'Order Date', 'Region', 'Category', 'Sub-Category',
                    'Product Name', 'Sales', 'Quantity', 'Discount', 'Profit']])
```

```python
# Check relationship between discount level and profit
import pandas as pd
discount_profit = df.groupby('Discount')['Profit'].mean().sort_values()
print(discount_profit)
```
